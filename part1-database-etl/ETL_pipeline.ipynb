# Pre-Processing : customers_raw data

```python
# Overview of columns and data types
df.info()
# NaN Check
df.isna().sum()
# Remove Duplicate entried if Any
df.drop_duplicates
# FIlling the email value 
df['email']=df['email'].fillna('Unknown')

#Standardizing the Phone numbers.

#Define the cleaning function
def format_india_phone(phone_str):
    if pd.isna(phone_str) or str(phone_str).strip() == "":
        return None
    
    try:
        # Convert to string and parse (Assuming 'IN' for India)
        parsed_number = ph.parse(str(phone_str), "IN")
        
        # Check if the number is actually valid according to Indian rules
        if ph.is_valid_number(parsed_number):
            # Format to E.164 (+91XXXXXXXXXX)
            return ph.format_number(
                parsed_number, 
                ph.PhoneNumberFormat.E164
            )
        else:
            return "Invalid" 
    except Exception:
        return "Error"


df['phone'] = df['phone'].apply(format_india_phone)

#Standardizing the date to YYYY-MM-DD format.

df['registration_date'] = pd.to_datetime(df['registration_date'], format='mixed', errors='coerce')
df['registration_date'] = df['registration_date'].dt.strftime('%Y-%m-%d')


#  Strip only the 'C' from the left side of the customer ID
df['customer_id'] = df['customer_id'].astype(str).str.lstrip('C')

# Making it an Integer
df['customer_id'] = pd.to_numeric(df['customer_id'], errors='coerce')

df['customer_id'] = df['customer_id'].astype(int)

unique_columns = ['first_name', 'last_name', 'customer_id']
df.drop_duplicates(subset=unique_columns, keep='first', inplace=True)

df_customers = df.copy()
df_customers
```

# Pre-Processing : products_raw data.

```python
#  Fill NaN with 0 for Stock Qty
df1['stock_quantity'] = df1['stock_quantity'].fillna(0)

# Convert to Title Case
df1['category'] = df1['category'].str.title()

# Filling the NaN in price using Average value grouping by Category.

df1['price'] = df1['price'].fillna(df1.groupby('category')['price'].transform('mean'))
df1['price'] = df1['price'].round(2)


# Synchronize the Orders table
df1['product_id'] = df1['product_id'].astype(str).str.lstrip('P')
df1['product_id'] = pd.to_numeric(df1['product_id'], errors='coerce')
df1['product_id'] = df1['product_id'].astype(int)

df_products = df1.copy()
df_products
```

# Pre-Processing : sales_raw data.

```python
#Standardizing the date to YYYY-MM-DD format.

df2['transaction_date'] = pd.to_datetime(df2['transaction_date'], format='mixed', errors='coerce')
df2['transaction_date'] = df2['transaction_date'].dt.strftime('%Y-%m-%d')

# Dropped the records without a customer ID as I noted that the trasaction status was pending which can skew my further queries outcomes.
df2.dropna(subset=['customer_id'], inplace=True)

# Filling the missing Product ID's using their Price in Products data.
price_to_id_map = dict(zip(df1['price'], df1['product_id']))

df2['product_id'] = df2['product_id'].fillna(df2['unit_price'].map(price_to_id_map))

# Dropping duplicate transactions.
unique_columns = ['transaction_date', 'customer_id', 'product_id']
df2.drop_duplicates(subset=unique_columns, keep='first', inplace=True)

# Synchronize the Orders table
df2['product_id'] = df2['product_id'].astype(str).str.lstrip('P')
df2['product_id'] = pd.to_numeric(df2['product_id'], errors='coerce')
df2['product_id'] = df2['product_id'].astype(int)

# Synchronize the Orders table
df2['customer_id'] = df2['customer_id'].astype(str).str.lstrip('C')
df2['customer_id'] = pd.to_numeric(df2['customer_id'], errors='coerce')
df2['customer_id'] = df2['customer_id'].astype(int)

# Synchronize the Orders table
df2['transaction_id'] = df2['transaction_id'].astype(str).str.lstrip('T')
df2['transaction_id'] = pd.to_numeric(df2['transaction_id'], errors='coerce')
df2['transaction_id'] = df2['transaction_id'].astype(int)

df2['total_amount'] = df2['unit_price'] * df2['quantity']

# Rename everything to match the schema exactly
df2 = df2.rename(columns={
    'transaction_id': 'order_id',
    'transaction_date': 'order_date',
    'subtotal': 'total_amount'
})

# Keep only the columns that actually exist in the 'orders' table
Order_schema_columns = ['order_id', 'customer_id', 'order_date', 'total_amount', 'status']
df_orders = df2[Order_schema_columns]

df2 = df2.rename(columns={'total_amount' : 'subtotal'})

Orderitems_schema_columns = ['order_id', 'product_id', 'quantity', 'unit_price','subtotal']
df_ordersitems = df2[Orderitems_schema_columns]

df2
```

# Uploading the Cleaned data into the SQL DataBase.


```python
from sqlalchemy import create_engine, MetaData, Table, Column, Integer, String, DECIMAL, Date, ForeignKey

# Creates 'fleximartData.db'
engine = create_engine('sqlite:///fleximartData.db')
metadata = MetaData()

# Define Tables
customers = Table(
    'customers', metadata,
    Column('customer_id', Integer, primary_key=True, autoincrement=True),
    Column('first_name', String(50), nullable=False),
    Column('last_name', String(50), nullable=False),
    Column('email', String(100), unique=False, nullable=False),
    Column('phone', String(20)),
    Column('city', String(50)),
    Column('registration_date', Date)
)
products = Table(
    'products', metadata,
    Column('product_id', Integer, primary_key=True, autoincrement=True),
    Column('product_name', String(100), nullable=False),
    Column('category', String(50), nullable=False),
    Column('price', DECIMAL(10, 2), nullable=False),
    Column('stock_quantity', Integer, default=0)
)
orders = Table(
    'orders', metadata,
    Column('order_id', Integer, primary_key=True, autoincrement=True),
    Column('customer_id', Integer, ForeignKey('customers.customer_id'), nullable=False),
    Column('order_date', Date, nullable=False),
    Column('total_amount', DECIMAL(10, 2), nullable=False),
    Column('status', String(20), default='Pending')
)
order_items = Table(
    'order_items', metadata,
    Column('order_item_id', Integer, primary_key=True, autoincrement=True),
    Column('order_id', Integer, ForeignKey('orders.order_id'), nullable=False),
    Column('product_id', Integer, ForeignKey('products.product_id'), nullable=False),
    Column('quantity', Integer, nullable=False),
    Column('unit_price', DECIMAL(10, 2), nullable=False),
    Column('subtotal', DECIMAL(10, 2), nullable=False)
)

# Create all tables in the database
metadata.create_all(engine)

df_customers.to_sql('customers', engine, if_exists='append', index=False)
df_products.to_sql('products', engine, if_exists='append', index=False)
df_orders.to_sql('orders', engine, if_exists='append', index=False)
df_ordersitems.to_sql('order_items', engine, if_exists='append', index=False)
```
